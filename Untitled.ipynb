```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('fenbu.csv')  # 替换为你的CSV文件路径

# 由于每一列的数据点个数不一样，我们需要将数据转换为长格式
df_long = df.melt(var_name='Element', value_name='Migration Barrier')

# 删除缺失值
df_long = df_long.dropna()

# 绘制箱线图
plt.figure(figsize=(10, 8))
sns.boxplot(x='Element', y='Migration Barrier', data=df_long, order=['Na', 'K', 'Rb', 'Ag', 'In'], linewidth=2.5)
plt.title('Migration Barrier Comparison (Box Plot)',fontsize=16)
plt.ylabel('Migration Barrier',fontsize=16)

plt.tick_params(axis='both', which='both', 
               direction='out', 
               length=5,   # 更长的刻度线
               width=1,    # 较粗的刻度线
               color='black',  # 刻度线颜色为黑色
               labelsize=16)

plt.gca().spines['top'].set_linewidth(1.2)   # 顶部框线
plt.gca().spines['right'].set_linewidth(1.2)  # 右侧框线
plt.gca().spines['bottom'].set_linewidth(1.2) # 底部框线
plt.gca().spines['left'].set_linewidth(1.2)   # 左侧框线
plt.show()

# 绘制小提琴图
plt.figure(figsize=(10, 8))
sns.violinplot(x='Element', y='Migration Barrier', data=df_long, order=['Na', 'K', 'Rb', 'Ag', 'In'],linewidth=2)
plt.title('Migration Barrier Comparison (Violin Plot)',fontsize=16)
plt.ylabel('Migration Barrier',fontsize=16)
plt.tick_params(axis='both', which='both', 
               direction='out', 
               length=5,   # 更长的刻度线
               width=1,    # 较粗的刻度线
               color='black',  # 刻度线颜色为黑色
               labelsize=16)
plt.gca().spines['top'].set_linewidth(1.2)   # 顶部框线
plt.gca().spines['right'].set_linewidth(1.2)  # 右侧框线
plt.gca().spines['bottom'].set_linewidth(1.2) # 底部框线
plt.gca().spines['left'].set_linewidth(1.2)   # 左侧框线
plt.show()
```
